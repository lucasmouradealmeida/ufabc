```python
#Import
import numpy as n
import math as m
import matplotlib.pyplot as plt
from scipy import *
from sympy import *
```

```python
#Exemplo 1 - Vibração Livre Não Amortecida - Considere a viga mostrada na Figura 8. Supõe-se que aproximadamente 1/4 da massa de 625 kg da viga esteja aplicada no ponto A. 
#Determine a frequência e o período de vibração e trace o gráfico.

#Dimensões e Propriedades da Viga
L = 10
b = 0.2
h = 0.5
E = 20*10**9
rho = 2500
M = 625

I = (b*(h**3))/12
K = (3*E*I)/(L**3)

#Condições iniciais
u0 = 100*10**-6
u0_ponto = 0.002

#frequências
w = m.sqrt(K/(M))
f = w/(2*m.pi)
T = 1/f

#Amplitude
A = m.sqrt((u0**2) + ((u0_ponto/w)**2))

#Ângulo de fase
theta = n.arctan((-u0_ponto)/(w*u0))

#função deslocamento
def u(t):
  return A*n.cos(w*t+theta)

#plot
y = []
t = n.arange(0, 5, 0.0001)

for i in t:
  y.append(u(i))

plt.plot(t,y)
plt.show()

```

```python
#Exemplo 2 - Vibração Livre Amortecida - Reutilizando a viga do Exemplo 1 e considerando uma taxa de amortecimento igual a 5%
#calcule a frequência amortecida e trace o gráfico do sistema livre e amortecido.

#Dimensões e Propriedades da Viga
L = 10
b = 0.2
h = 0.5
E = 20*10**9
rho = 2500
M = 625
e = 5/100

I = (b*(h**3))/12
K = (3*E*I)/(L**3)

#Condições iniciais
u0 = 100*10**-6
u0_ponto = 0.002

#frequências
w = m.sqrt(K/(M))
f = w/(2*m.pi)
T = 1/f
wd = w*m.sqrt(1-e**2)

#Amplitude
A = m.sqrt((u0**2) + (((u0_ponto+e*w*u0)/wd)**2))

#Ângulo de fase
theta = -n.arctan((u0_ponto+e*w*u0)/(wd*u0))

#função deslocamento
def u(t):
  return m.exp(-e*w*t) * A * n.cos(wd*t+theta)

#plot
y = []
t = n.arange(0, 5, 0.0001)

for i in t:
  y.append(u(i))

plt.plot(t,y)
plt.show()






```

```python
#Exmeplo 3 - Carregamento Harmônico - Considere novamente a viga dos Exemplos 1 e 2. No ponto A é instalado um motor com massa 500 kg que 
#aplica uma força dinâmica igual a p(t) = 500*sen(105*t) N. Determine ω, ωD, u, u̇ e ü no regime estacionário.

#Dimensões e Propriedades da Viga
L = 10
b = 0.2
h = 0.5
E = 20*10**9
rho = 2500
M = 625+500
e = 5/100

I = (b*(h**3))/12
K = (3*E*I)/(L**3)

#Condições iniciais
u0 = 100*10**-6
u0_ponto = 0.002


#Solução Homogênea

#frequências
w = m.sqrt(K/(M))
f = w/(2*m.pi)
T = 1/f
wd = w*m.sqrt(1-e**2)

#Amplitude
A = m.sqrt((u0**2) + (((u0_ponto+e*w*u0)/wd)**2))

#Ângulo de fase
theta = -n.arctan((u0_ponto+e*w*u0)/(wd*u0))

#Solução Particular
omega = 105
p0 = 500
beta = omega/w
Ap = (p0/K)*(1/m.sqrt((1-beta**2)**2+(2*e*beta)**2))
thetap = n.arctan((2*e*beta)/(1-beta**2))

#função deslocamento
def u(t):
  return (m.exp(-e*w*t) * A * n.cos(wd*t+theta)) + (Ap*n.sin(omega*t+thetap))

def uh(t):
  return (m.exp(-e*w*t) * A * n.cos(wd*t+theta))

def up(t):
  return (Ap*n.sin(omega*t+thetap))

#plot
y = []
yh = []
yp = []
t = n.arange(0, 5, 0.0001)

for i in t:
  y.append(u(i))

for i in t:
  yh.append(uh(i))

for i in t:
  yp.append(up(i))

plt.plot(t,y,label='u(t)')
plt.plot(t,yh,label='uh(t)')
plt.plot(t,yp,label='up(t)')
plt.legend()
plt.show()

#Derivatives

x = Symbol('x')
yd = Ap*sin(omega*x+thetap)
yd_ponto = yd.diff(x)
print(yd_ponto)
yd_2pontos = yd_ponto.diff(x)
print(yd_2pontos)

```
