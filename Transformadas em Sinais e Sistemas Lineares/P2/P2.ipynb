```python
#IMPORTS
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
Hvw = lambda w: (1j*w)/((1j*w)**2 + (3*1j*w) + (2))
```

```python
ws = np.linspace(0,100,100)
plt.plot(ws,np.abs(Hvw(ws)))
plt.grid()
plt.xlabel('omega')
plt.ylabel('|Hv(jw)|')
plt.title('Espectro de Amplitude')
plt.show()
```

```python
plt.plot(ws,np.angle(Hvw(ws)))
plt.grid()
plt.xlabel('omega')
plt.ylabel('<Hv(jw)')
plt.title('Espectro de Fase')
plt.show()
```
