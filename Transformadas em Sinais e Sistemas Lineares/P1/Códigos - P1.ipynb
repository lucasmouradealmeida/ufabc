```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from sympy import *
```

```python
#Prova P1 - Valores Iniciais

R = 0.8
C = 0.1
A = 2
T0 = 5
TH = 2
TL = 3
t = np.linspace(0,10,1000)
j = 0+1j
w0 = 2*np.pi/T0
```

```python
#3) Gráfico do Sinal v(t) - Forma trigonométrica
x0 = 2*TH/T0
xk = lambda k: (1-(np.exp(-j*k*4*np.pi/5)))/(j*k*np.pi)

ak = lambda k: 2*np.real((1-(np.exp(-j*k*4*np.pi/5)))/(j*k*np.pi))
bk = lambda k: -2*np.imag((1-(np.exp(-j*k*4*np.pi/5)))/(j*k*np.pi))

v_t = x0 #valor inicial
N = 5000
for k in range(1,N):
  v_t = v_t + ak(k)*np.cos(k*w0*t) + bk(k)*np.sin(k*w0*t)

plt.figure()
plt.plot(t,v_t,linewidth=3,label='v(t)')
plt.grid(True)
plt.xlabel('Tempo')
plt.legend()
plt.title('N = %i' %N)
```

```python
#4) v(t) - Gráfico dos espectros de amplitude e de fase - Forma Compacta
c0 = 4/5
ck = lambda k: (2/(k*np.pi))*(np.sqrt(2*(1-np.cos(k*4*np.pi/5))))
theta = lambda k: -1*np.arctan((1-np.cos(k*4*np.pi/5))/(np.sin(k*4*np.pi/5)))

N=20
t = np.arange(0,N,1)
v_fase = np.zeros((N,1))
v_amp = np.zeros((N,1))

v_amp[0,0]= c0

for k in range(1,N):
  v_amp[k,0]= ck(k)
  v_fase[k,0]= theta(k)

#Grafico Fase v(t)
plt.figure()
markerline1, stemlines, _ = plt.stem(t,v_fase[:,0],label='v(t)')
plt.setp(markerline1,'markerfacecolor','b')
plt.setp(stemlines,'linewidth',3)
plt.title('Espectro de fase')
plt.legend()
plt.grid(True)
plt.xticks(t)
plt.xlabel('N')
plt.yticks([-np.pi/2,0,np.pi/2],('$-\pi/2$','$0$','$\pi/2$'))
plt.show()

#Gráfico Amplitude v(t)
plt.figure()
markerline1, stemlines, _ = plt.stem(t,v_amp[:,0],label='v(t)')
plt.setp(markerline1,'markerfacecolor','b')
plt.setp(stemlines,'linewidth',3)
plt.title('Espectro de amplitude')
plt.legend()
plt.grid(True)
plt.xticks(t)
plt.xlabel('N')
plt.show()
```
